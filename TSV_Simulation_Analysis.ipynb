```python
# Install library quietly
!pip install pythermalcomfort -q

import pandas as pd
from pythermalcomfort.models import pmv_ppd_ashrae
from pythermalcomfort.utilities import v_relative, clo_dynamic_ashrae
from google.colab import files
import numpy as np

# Load the uploaded file (assumes 'integrated_13_institutions_168h.xlsx')
filename = 'integrated_13_institutions_168h.xlsx'
print(f"Processing file: {filename}")
df = pd.read_excel(filename)

# Extract hour from datetime column
df['hour'] = df['datetime'].dt.hour

# PMV calculation function (met=1.0, dynamic clo: 0.45 daytime, 0.75 nighttime)
def calculate_perfect_tsv(row):
    try:
        ta = row['TA_mean']
        tr = row['TG_mean']
        rh = row['RH_mean']
        vel = row['Average_Wind_Speed_m_s'] # Changed to renamed column name
        met = 1.0

        if pd.isna(ta) or pd.isna(tr) or pd.isna(rh) or pd.isna(vel):
            return np.nan

        clo = 0.75 if (row['hour'] < 7 or row['hour'] >= 23) else 0.45
        v_r = v_relative(v=vel, met=met)
        clo_d = clo_dynamic_ashrae(clo=clo, met=met)

        results = pmv_ppd_ashrae(tdb=ta, tr=tr, vr=v_r, rh=rh, met=met, clo=clo_d, model="55-2023")
        return round(results.pmv, 3)
    except Exception as e:
        # print(f"Error in row {row.name}: {e}") # Removed print statement
        return np.nan

# Compute Perfect_TSV
df['Perfect_TSV'] = df.apply(calculate_perfect_tsv, axis=1)

# Summary statistics
print("\n" + "="*60)
print("Guiyang Elderly Care Thermal Comfort Results")
print("="*60)
print(f"Overall Perfect_TSV Mean:     {df['Perfect_TSV'].mean():.3f}")
print(f"Nighttime (00-06) Mean:      {df[df['hour']<7]['Perfect_TSV'].mean():.3f}")
print(f"Daytime (07-22) Mean:        {df[(df['hour']>=7) & (df['hour']<23)]['Perfect_TSV'].mean():.3f}")
print(f"Comfort Rate (-0.5 to +0.5): {((df['Perfect_TSV'] >= -0.5) & (df['Perfect_TSV'] <= 0.5)).mean()*100:5.1f}%")
print(f"Cold Discomfort (< -0.5):    {(df['Perfect_TSV'] < -0.5).mean()*100:4.1f}%")
print(f"Heat Discomfort (> +0.5):    {(df['Perfect_TSV'] > +0.5).mean()*100:4.1f}%")
print("="*60)

# Save and download
output_file = 'Perfect_TSV_Guiyang_Elderly_13Institutions.xlsx'
df.to_excel(output_file, index=False)
files.download(output_file)

print(f"\nDownloaded: {output_file}")
print("Use Perfect_TSV for all analyses (SCI papers, ML training, SHAP).")
```
